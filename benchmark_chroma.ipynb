```python
from chroma_research import GeneralBenchmark
from sentence_transformers import SentenceTransformer
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
chunker = RecursiveCharacterTextSplitter()
```

```python
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction, DefaultEmbeddingFunction

default = DefaultEmbeddingFunction()
snowflake_s = SentenceTransformerEmbeddingFunction(model_name="Snowflake/snowflake-arctic-embed-s", device="mps")
snowflake_m = SentenceTransformerEmbeddingFunction(model_name="Snowflake/snowflake-arctic-embed-m", device="mps")
```

```python
benchmark = GeneralBenchmark()
```

```python
results_default = benchmark.run(chunker=chunker, embedding_function=default)
print(results_default)
```

```python
results_snowflake_s = benchmark.run(chunker=chunker, embedding_function=snowflake_s)
print(results_snowflake_s)
```

```python
results_snowflake_m = benchmark.run(chunker=chunker, embedding_function=snowflake_m)
print(results_snowflake_m)
```
